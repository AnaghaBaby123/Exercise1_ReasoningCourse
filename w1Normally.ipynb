Applying Box-Muller transform to generate normally distributed data from a uniformly distributed data space

```python
a = 4514785.02546
b = 14535214.2546
r = 21458125145213
m = 100  # decides the upper limit of the random numbers generated the value can be between 0 and 100

def random():
    global r
    r = (a*r + b)%m
    return r
```

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

def generate_normal():
    r1 = random()
    r2 = random()
    new_r1 = r1/100 #To make the value in the range (0,1)
    new_r2 = r2/100 #To make the value in the range (0,1)
    z0 = math.sqrt(-2* math.log(new_r1)) * math.cos(2*math.pi*new_r2) #Applying box muller transform
    z1 = math.sqrt(-2* math.log(new_r1)) * math.sin(2*math.pi*new_r2) 
    return z0,z1
    


norm_nos =[]

for i in range(1000):
    norm_nos.append(generate_normal()[0])
    norm_nos.append(generate_normal()[1])
print(norm_nos)
```

```python
plt.hist(norm_nos,)
plt.show()
```

Calculating mean and standard deviation

```python
sums=0
for i in range(len(norm_nos)):
    sums = sums + norm_nos[i]
mean =sums/len(norm_nos)
print("Mean=",mean)
```

```python
var = 0
sums = 0
for i in range(len(norm_nos)):
    sums = sums + (norm_nos[i] - mean)**2
var = sums/ len(norm_nos)
sd = math.sqrt(var)
print("Standard deviation:",sd)
    
```

```python
# Plot histogram with KDE plot
sns.histplot(norm_nos, kde=True)

# Add markers for mean and standard deviation
plt.axvline(x=mean, color='r', linestyle='--', label='Mean')
plt.axvline(x=mean + sd, color='g', linestyle='--', label='Mean + Std Dev')
plt.axvline(x=mean - sd, color='g', linestyle='--', label='Mean - Std Dev')

# Add legend
plt.legend()

# Show plot
plt.show()
```
