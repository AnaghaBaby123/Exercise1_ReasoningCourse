# 2 a) Generation of dataset:

```python
a = 4514785.02546
b = 14535214.2546
r = 21458125145213
m = 100  # decides the upper limit of the random numbers generated the value can be between 0 and 100

def random():
    global r
    r = (a*r + b)%m
    return r
```

```python
print(random())
```

```python
dataset = []

for i in range(1000000):
    r = int(random())
    if(dataset.count(r)<1100):
        dataset.append(r)
print(dataset)
print(len(dataset))
```

```python
#find the frequencies 

number_dict ={}
for i in dataset:
    if i not in number_dict.keys():
        number_dict[i] = 1
    else:
        number_dict[i] = number_dict[i] +1


sorted_number_dict = dict(sorted(number_dict.items()))
print(sorted_number_dict)
```

```python
#plotting using matplotlib


import matplotlib.pyplot as plt
plt.plot(sorted_number_dict.keys(),sorted_number_dict.values(), color='red')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

```python
#Calculating the parameters: Mean and standard deviation
import math
nos =[]
for i in sorted_number_dict.keys():
    nos.append(i)
a = nos[0]
b = nos[len(nos)-1]


m = (a + b)/2
print("Mean:", m)

sd = math.sqrt(((b-a)*(b-a))/12)
print("standard deviation:",sd)
```

# 2 b) Sampling 1000:

```python
sample =[]
for i in range(1000):
    r = int(random())
    sample.append(dataset[r])
print(sample)
```

```python
#finding the frequencies
sample_number_dict ={}
for i in sample:
    if i not in sample_number_dict.keys():
        sample_number_dict[i] = 1
    else:
        sample_number_dict[i] = sample_number_dict[i] +1


sorted_sample_number_dict = dict(sorted(sample_number_dict.items()))
print(sorted_sample_number_dict)
```

```python
#plotting using matplotlib


import matplotlib.pyplot as plt

plt.scatter(sorted_sample_number_dict.keys(),sorted_sample_number_dict.values(),color='blue')
plt.show()
```

```python
#Calculating the parameters: Mean and standard deviation
import math
nos =[]
for i in sorted_sample_number_dict.keys():
    nos.append(i)
a = nos[0]
b = nos[len(nos)-1]


m = (a + b)/2
print("Mean:", m)

sd = math.sqrt(((b-a)*(b-a))/12)
print("standard deviation:",sd)
```

We observe the mean and standard deviation  are same before and after the sampling.
