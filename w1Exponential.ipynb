Inverse transform method

To generate exponentially distributed random numbers from random numbers, we use the inverse transform method

```python
a = 4514785.02546
b = 14535214.2546
r = 21458125145213
m = 100  # decides the upper limit of the random numbers generated the value can be between 0 and 100

def random():
    global r
    r = (a*r + b)%m
    return r
```

```python
print(random())
```

```python

import math
import matplotlib.pyplot as plt

def generate_exponential(u,lambda_val):
    x = - (1 / lambda_val) * math.log(1 - u)  # Apply inverse transform formula
    return x


lambda_val = 0.5  # Rate parameter of the exponential distribution
exp_nos =[]
nos=[]
for i in range(1000):  # creating a random numbers of 1000 samples.
    r = random()
    new_r = r/100 #To make the value in the range (0,1)
    nos.append(new_r)
for i in range(len(nos)):
    exp_nos.append(generate_exponential(nos[i],lambda_val))
#print(exp_nos)
```

```python
plt.hist(exp_nos, bins=40, alpha=0.5, color='g')
```

```python
#Calculation mean and sd
mean = 1/lambda_val
sd = 1/lambda_val
print("Mean is",mean)
print("Standard deviation is",sd)
```
